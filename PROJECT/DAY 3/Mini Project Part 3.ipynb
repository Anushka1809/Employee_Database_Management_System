```python
def check_hr(id):
    flag = False
    file = open("hr.txt", 'r')
    hr = file.readlines()
    for i in hr:
        j = i.split('  ')
        if id == j[0]:
            flag = True
    return flag


def view_details(id):
    file = open("employee.txt", 'r')
    emp = file.readlines()
    for i in emp:
        j = i.split('  ')
        if id == j[0]:
            return i
    else:
        return "Employee doesn't exist"


def view_employee(designation):
    file = open('employee.txt', 'r')
    emp = file.readlines()
    if(designation=='all'):
       # file2 = open('employee.txt', 'r')
       # x = file2.readlines()
        for i in emp:
            print(i)
            
    else:
        for i in emp:
            flag=1
            j = i.split('  ')
            if designation == j[3]:
                print(i)
        if flag==0:
            print("not found")
                    
    return 0


def view_hr():
    file1 = open('hr.txt', 'r')
    file2 = open('employee.txt', 'r')
    hr = file1.readlines()
    emp = file2.readlines()
    for i in hr:
        for j in emp:
            k = i.split('  ')
            l = j.split('  ')
            if k[0] == l[0]:
                print(k[1] + ',' + l[1] + ',' + k[2])
                continue
    return 0


__name__ == 'main'
id = input("Enter the id again: ")
if check_hr(id):
    #if input is an hr
    while 1:
        print("Enter 1 to view own details\nEnter 2 to view All employees\nEnter q to exit")
        c = input("Enter your Option->  ")
        if c == '1':
            print(view_details(id))
        elif c == '2':
            designation = input("Enter designation: ")
            view_employee(designation)
        elif c == 'q':
            break
else:
    while 1:
        #if input is not an hr
        print("Enter 1 to view own details\nEnter 2 to view all HR names\nEnter q to exit")
        c = input("Enter your Option->  ")
        if c == '1':
            print(view_details(id))
        elif c == '2':
            view_hr()
        elif c == 'q':
            break
            
```
